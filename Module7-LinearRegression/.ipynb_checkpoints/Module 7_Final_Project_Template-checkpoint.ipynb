### 2.3 Data Preparation

Determine if there are any missing values in the data. Did the data need to be reshaped? If yes, include a description of the steps you followed to clean the data.

##### Response

I imputed the frontage area of 14 properties which were null by creating a ratio of the average frontage area to the average lot area and multiplying that by that ratio to the lot areas of the properties missing frontage area data. I tried both ways and this imputation method was slightly more accurate than simply using the average of all frontage areas.

The remaining columns with missing values did not have a garage, fireplace, or fireplace and needed to have the blanks filled with NA values except GarageYrBlt. The YearBuilt was used to fill GarageYrBlt because there were only 6 NAs for the year the garage was built and the buildings were relatively older and lower valued. It was important to have a numerical value for this field because I wanted to included it in the lin
Every row in the dataset was missing pool data, so PoolArea and PoolQC were dropped.

```python
pd.options.display.max_rows = 200
nanCount = data.isnull().sum()
nanCount = nanCount[(nanCount!=0)]
print(nanCount.sort_values(ascending = False))

def cleanData(df):
    haveFrontage = df.dropna(subset=['LotFrontage'])
    avgFrontage = df['LotFrontage'].mean()
    avgLotArea = df['LotArea'].mean()
    frontageAreaRatio = avgFrontage/avgLotArea
    df.LotFrontage.fillna(df.LotArea * frontageAreaRatio, inplace = True)
    df.Alley.fillna('NA', inplace = True)
    df.Fence.fillna('NA', inplace = True)
    df.GarageType.fillna('NA', inplace = True)
    df.GarageYrBlt.fillna(df['YearBuilt'], inplace = True)
    df.GarageFinish.fillna('NA', inplace = True)
    df.GarageQual.fillna('NA', inplace = True)
    df.GarageCond.fillna('NA', inplace = True)
    df.FireplaceQu.fillna('NA', inplace = True)
    df.MasVnrType.fillna('None', inplace = True)
    df.MiscFeature.fillna('NA', inplace = True)
    df.BsmtQual.fillna('NA', inplace = True)
    df.BsmtCond.fillna('NA', inplace = True)
    df.BsmtExposure.fillna('NA', inplace = True)
    df.BsmtFinType1.fillna('NA', inplace = True)
    df.BsmtFinType2.fillna('NA', inplace = True)
    df.drop(columns = ['PoolArea', 'PoolQC'], inplace = True, errors = 'ignore')
    return df
data = cleanData(data)
```

### 3.1 Linear Regression

Give a description (500 or less words) of the algorithm you use in this project. Include mathematical and computational details about linear regression.

Include details about the theory (origin of the method, derivation, and formulas) and the necessary steps to implement the algorithm using Python.

###### Response

Linear Regression is a statistical method to predict the value of a dependent variable based on one or several independent variables. The technique was first published by Legendre in 1805 and by Gauss in 1809, although Gauss claimed to have known of the method since 1795. The simplest form of linear regression, using only one variable is the same as the equation for a line, y = mx + b with x as the independent variable, y as the dependent variable, m as the slope, and b as the y-intercept. However, there can be several dependent variables and slope coefficients as seen here: Y = 0 + 1*X1 + 2*X2 +  + n*Xn with X1, X2,  Xn as the dependent variables, and 1, 2,  n as their coefficients.

The python implementation of linear regression we learned in this course was from the scikit learn package. The documentation states it is an "ordinary least square Linear Regression" which minimizes the residual sum of squares (RSS) between the observed target in the dataset and the targets predicted by the linear approximation. To implement the algorithm in Python we must generate a model and then train the model on our dependent and independent variables using the fit() function. When that is complete, we can use the model to make predictions on the dataset again and generate a score. The difference between the actual and predicted outcomes and score can help us assess how good our model is.

When deciding to use a linear regression model, it is important to consider if the data is suited to that type of regression. The assumption about data suitable for analysis with linear regression are:
* Continuous variables
* Linear relationship between the variables
* No dependency between the variables, the are independent of each other
* No significant outliers
* Homoscedasticity - variances along the best-fit linear-regression line remain similar all through that line
* Residual errors of the best-fit regression follow a normal distribution

### 3.2 Analysis 

Implement the algorithm on your data according to the examples in Video 7.9 and Video 7.10.

Try to improve the results of your model analysis by including a different number of variables in your code for linear regression. Use what you learned about the correlation between variables when you explored your data to help you select these variables.

Compare the results of at least three different groups of variables. In other words, run a linear regression algorithm on at least three different sets of independent variables. How many variables to include in each set is up to you.

For each step, make sure you include your code. Ensure that your code is commented.

```python
# a function which transforms columns named in colsToLog
def transformData(df):
    colsToLog = ['LotArea', 'OverallQual', '1stFlrSF', 'GrLivArea', 'TotRmsAbvGrd'] # columns which should be logged
    df[colsToLog] = df[colsToLog].apply(np.log) # take the log of columns which were found to be more symmetrical when logged
    return df
```

```python
# model 1 - a model which incorporates logs
cols = corrData.index[1:17] # select columns to include in the model

X1 = data[cols] # create a copy of the independent variables we want to include in the model
X1 = transformData(X1)
Y1 = np.log(data['SalePrice']) # log of the sale price we are predicting

# generate the model and predictions
lr = linear_model.LinearRegression()
model1 = lr.fit(X1,Y1)
predictions1 = model1.predict(X1)
```

```python
# model 2 - a model which does not incorporate logs to be compared to model 1
X2 = data[cols] # create a copy of the independent variables we want to include in the model
Y2 = data['SalePrice'] # do not take the log of the sale price for model 2

# generate the model and predictions
lr = linear_model.LinearRegression()
model2 = lr.fit(X2,Y2)
predictions2 = model2.predict(X2)
```

```python
# model 3 - a model for the things I would estimate would predict price best, rather than tinkering with the model to optimize score
cols3 = ['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath',
        '1stFlrSF', 'YearRemodAdd', 'LotArea', '2ndFlrSF','LotFrontage', 'BedroomAbvGr', 
        'MSSubClass','OverallCond'] # select columns to include in the model
X3 = data[cols3] # create a copy of the independent variables we want to include in the model
Y3 = np.log(data['SalePrice'])

# generate the model and predictions
lr = linear_model.LinearRegression()
model3 = lr.fit(X3,Y3)
predictions3 = model3.predict(X3)
```
