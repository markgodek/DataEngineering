### 2.1 Import the Data

I imported the data from a file on my PC and stored it in a variable.
I printed the head of the data to ensure it was properly read.
The file appears to contain 100 observations of home sales with 82 columns describing the home sales.

```python
import pandas as pd
import matplotlib.pyplot as plt

# store the path to the data in a string and use pandas to read the data from the path
path = r'C:\Users\markg\OneDrive\Data Engineering Course\Housing\houseSmallData.csv'
data = pd.read_csv(path)
print(data.head)
```

[Back to top](#Index)

### 2.2 Data Exploration

I calculated the correlation of each column to the sale price of the homes and sorted them in descending order.

I created plots for the columns that were most correlated with the sale price.

Overall quality is discrete number from 1 to 10 that is the single best predictor of sale price.

GRLivArea is a continuous variable which describes the above grade living area in the home and is well correlated to sale price.

GarageArea describes how much garage area a home has. It is well correlated to sale price, although some homes without garages sell for higher prices than some with even moderately sized garages. The graph which follows this is GarageCars, which likely recapitulates the same results as GarageArea, but with a discrete variable for how many cars a garage fits.

The final of the correlations I chose to visualize is YearBuilt. There is a slightly positive correlation that the newer the home, the higher the sale price.


```python
corr_data = data.corr(numeric_only = True)['SalePrice'].sort_values(ascending = False)
print(corr_data)
for x in corr_data.index[1:6]:

    plt.scatter(data[x], data.SalePrice)
    plt.title(x)
    ax = plt.gca() # get matplotlib axes
    plt.text(.025, .95, "Correlation coefficient ={:.2f}".format(corr_data[x]), transform=ax.transAxes)
    plt.show()
```

### 2.3 Data Preparation

Determine if there are any missing values in the data. Did the data need to be reshaped? If yes, include a description of the steps you followed to clean the data.

I imputed the frontage area of 14 properties which were null by creating a ratio of the average frontage area to the average lot area and multiplying that by that ratio to the lot areas of the properties missing frontage area data. I tried both ways and this imputation method was slightly more accurate than simply using the average of all frontage areas.

The remaining columns with missing values did not have a garage, fireplace, or fireplace and needed to have the blanks filled with NA values.
Every row in the dataset was missing pool data, so PoolArea and PoolQC were dropped.

```python
pd.options.display.max_rows = 200
nan_count = data.isnull().sum()
nan_count = nan_count[(nan_count!=0)]
print(nan_count.sort_values(ascending = False))

have_frontage = data.dropna(subset=['LotFrontage'])
avg_frontage = data['LotFrontage'].mean()
avg_lotarea = data['LotArea'].mean()
frontage_area_ratio = avg_frontage/avg_lotarea
data.LotFrontage.fillna(data.LotArea * frontage_area_ratio, inplace = True)
data.Alley.fillna('NA', inplace = True)
data.Fence.fillna('NA', inplace = True)
data.GarageType.fillna('NA', inplace = True)
data.GarageYrBlt.fillna('NA', inplace = True)
data.GarageFinish.fillna('NA', inplace = True)
data.GarageQual.fillna('NA', inplace = True)
data.GarageCond.fillna('NA', inplace = True)
data.FireplaceQu.fillna('NA', inplace = True)
data.MasVnrType.fillna('None', inplace = True)
data.MiscFeature.fillna('NA', inplace = True)
data.BsmtQual.fillna('NA', inplace = True)
data.BsmtCond.fillna('NA', inplace = True)
data.BsmtExposure.fillna('NA', inplace = True)
data.BsmtFinType1.fillna('NA', inplace = True)
data.BsmtFinType2.fillna('NA', inplace = True)
#data.drop(columns = ['PoolArea', 'PoolQC'], inplace = True)
```
